### Goal : Predict the percentage of an student based on the no. of study hours.

### 1. Importing Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import sklearn
```

### 2. Loading the dataset and pre-process the data.

```python
df = pd.read_csv('student_scores - student_scores.csv')
print(f'Dataset shape : {df.shape}\n')
print(df.head())
print(f'features :{df.columns}')
```

```python
df.plot(x='Hours', y='Scores', style='^')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X, y = df.iloc[:, :-1].values, df.iloc[:, 1].values
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 3. Train the data

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
print("Training Finished")
```

### 4. Visualization.

```python
line = lin_reg.coef_*X + lin_reg.intercept_
plt.scatter(x=X, y=y)  
plt.plot(X, line)
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

### 4. Evaluating the Model.

```python
from sklearn.metrics import mean_absolute_error 
y_pred = lin_reg.predict(x_test)
print(f'Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}')
```

```python
# What will be predicted score if a student studies for 9.25 hrs/ day? 
hours = [[9.25]] 
pred = lin_reg.predict(hours)
print(f'Number of hours:{9.25}, Predicted Score:{pred[0]}')
```
